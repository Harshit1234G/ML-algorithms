## On Iris dataset

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
data = iris.data
target = iris.target
```

```python
test_index = [0, 50, 100]
```

```python
train_features = np.delete(data, test_index, axis= 0)
train_labels = np.delete(target, test_index)
test_features = data[test_index]
test_labels = target[test_index]
```

```python
clf = DecisionTreeClassifier()
```

```python
clf.fit(train_features, train_labels)
```

```python
clf.predict(test_features)
```

## To visualize the Tree (clf)

```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(figsize=(28, 28))
plot_tree(
    clf, 
    feature_names= iris.feature_names,
    class_names= iris.target_names,
    filled= True
)
fig.savefig("Tree.png")
```
