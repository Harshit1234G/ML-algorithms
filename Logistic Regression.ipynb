## Train a logistic regression classifier to predict whether a flower is iris virginica or not

```python
from sklearn.datasets import load_iris
from sklearn.linear_model import LogisticRegression
import numpy as np
import matplotlib.pyplot as plt
```

```python
iris = load_iris()
```

```python
iris.data
```

```python
print(iris.DESCR)
```

```python
features = iris.data[:, 3:]
labels = (iris.target == 2).astype('int')
```

```python
labels, features
```

```python
classifier = LogisticRegression()
```

```python
classifier.fit(features, labels)
```

```python
new_features = np.linspace(0, 3, 1000).reshape(-1, 1)
```

```python
new_features
```

```python
prediction = classifier.predict_proba(new_features)
```

```python
prediction    
#first one is the probabilty of event not occuring and second one is the probabilty of occuring in this case as in provided labels first values are 0
```

```python
plt.plot(new_features, prediction[:, 1])
plt.xlabel('Feature')
plt.ylabel('Probability')
plt.title('Curve for the flower being a virginica')
```

```python
plt.plot(new_features, prediction[:, 0])
plt.xlabel('Feature')
plt.ylabel('Probability')
plt.title('Curve for the flower not being a virginica')
```

```python
plt.plot(new_features, prediction)
plt.xlabel('Feature')
plt.ylabel('Probability')
plt.title('Plot for both')
```

```python
classifier.predict_proba([[1]])
```
