```python
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
diabetes = datasets.load_diabetes()
```

```python
diabetes
```

```python
print(diabetes.DESCR)
```

```python
diabetes.data.shape
```

```python
diabetes.target.shape
```

```python
diabetes.data
```

```python
data = diabetes.data[:, 2].reshape(-1, 1)
```

```python
data
```

```python
diabetes_x_train = data[:-30]
diabetes_y_train = diabetes.target[:-30]
diabetes_x_test = data[-30:]
diabetes_y_test = diabetes.target[-30:]
```

```python
model = LinearRegression()
```

```python
model.fit(diabetes_x_train, diabetes_y_train)
```

```python
prediction = model.predict(diabetes_x_test)
```

```python
prediction
```

```python
mean_squared_error(diabetes_y_test, prediction)
```

```python
model.coef_, model.intercept_
```

```python
diabetes_x_test.shape, diabetes_y_test.shape
```

```python
plt.scatter(diabetes_x_test, diabetes_y_test)
```

```python
plt.scatter(diabetes_x_test, prediction)
```

```python
plt.scatter(diabetes_x_test, diabetes_y_test)
plt.plot(diabetes_x_test, prediction)
```
